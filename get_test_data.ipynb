```python
from PIL import Image
from random import randint
import requests
import time
import os
```

```python
target_dir = "./test/"
```

```python
user_agent = "Saucer/0.0.0"
headers = {
    'user-agent': user_agent
}
```

```python
# Get a list of posts to download
posts = requests.get("https://e926.net/posts.json", headers=headers, params={
    'limit': 100,
    'tags': 'rating:safe filetype:png'
})
```

```python
# Get post URLs
urls = list(map(lambda post: post['file']['url'], posts.json()['posts']))
```

```python
for i, url in enumerate(urls):
    print(f"{i + 1} / {len(urls)}")
    print(f"Downloading {url}...")
    b = requests.get(url, headers=headers).content
    with open(target_dir + f"uncropped/{i}.png", "wb") as f:
        f.write(b)
    time.sleep(1)
```

```python
# Crop
for file in os.listdir(target_dir + "uncropped/"):
    print(file)
    img = Image.open(target_dir + "uncropped/" + file)
    crop_x = randint(0, img.width - 1)
    crop_y = randint(0, img.height - 1)
    crop_width = randint(1, img.width - crop_x)
    crop_height = randint(1, img.height - crop_y)
    img = img.crop((crop_x, crop_y, crop_x + crop_width, crop_y + crop_height))
    img.save(target_dir + "cropped/" + file)
```
